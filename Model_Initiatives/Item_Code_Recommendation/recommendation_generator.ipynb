```python
import pandas as pd 
import numpy as np 


item_code = pd.read_csv('cleaned_item_code.csv')
item_code
```

```python
item_code_non_extreme = item_code[item_code['price AUD'] <= 500]
```

```python
item_code_list = item_code_non_extreme['item_number'].tolist()

# do apply lambda for item_code['recommened_item'] being random choice of item_code_list which is not the same as item_code['item_code']
item_code['recommended_item'] = item_code['item_number'].apply(lambda x: np.random.choice([i for i in item_code_list if i != x]))

# make column 'recommended_item' as dtype object 
item_code['recommended_item'] = item_code['recommended_item'].astype('object')
```

```python
item_code[['item_number', 'recommended_item']].to_csv('recommendation_pair.csv', index=False)
```
