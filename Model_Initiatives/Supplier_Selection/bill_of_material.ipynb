```python
import pandas as pd
import numpy as np

# Parameters
item_code_cleaned = pd.read_csv('cleaned_item_code.csv')

codes = item_code_cleaned['item_number'].tolist()

materials_bank = [f"Material {i}" for i in range(1, 7)]  # Materials from "Material 1" to "Material 30", (1, 31)
quantity_range = (1, 10)  # Quantity range (1 to 20)

# Create dummy BoM dataset
data_list = []

for code in codes:
    num_materials = np.random.randint(1, 4)  # Randomly choose 3-5 materials per Code
    selected_materials = np.random.choice(materials_bank, num_materials, replace=False)
    quantities = np.random.randint(quantity_range[0], quantity_range[1] + 1, num_materials)
    
    for material, quantity in zip(selected_materials, quantities):
        data_list.append({"Code": code, "Material": material, "Quantity": quantity})

# Convert to a DataFrame
bom_df = pd.DataFrame(data_list)

# Display the dummy BoM dataset
bom_df
```

```python
bom_df.to_csv('bom_single_clinic.csv', index=False)
```
