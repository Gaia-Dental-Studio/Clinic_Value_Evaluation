```python
import pandas as pd 
import numpy as np 
import pickle 
```

```python
# load pickle file
with open('item_code_per_clinic.pkl', 'rb') as file:
    model = pickle.load(file)
    
yearly_demand_projection = {}

for clinic, dataframe in model.items():
    yearly_demand_projection[clinic] = dataframe[['Code', 'Total Demand']]
    
yearly_demand_projection
```

```python

# Ensure all 'Code' values are zero-padded to three digits
for clinic, df in yearly_demand_projection.items():
    yearly_demand_projection[clinic]["Code"] = df["Code"].astype(str).apply(lambda x: x.zfill(3))

# Initialize an empty list to store intermediate data
data_list = []

# Loop through the dictionary to extract data
for clinic, df in yearly_demand_projection.items():
    for _, row in df.iterrows():
        data_list.append({"Clinic": clinic, "Code": row["Code"], "Total Demand": row["Total Demand"]})

# Convert the data list into a DataFrame
long_df = pd.DataFrame(data_list)

# Pivot the DataFrame to create the demand projection matrix
demand_projection_matrix = long_df.pivot(index="Clinic", columns="Code", values="Total Demand")

# Fill NaN with 0 if there are missing demand values
demand_projection_matrix = demand_projection_matrix.fillna(0)
```

```python
demand_projection_matrix.to_csv('demand_projection_matrix.csv')
```
